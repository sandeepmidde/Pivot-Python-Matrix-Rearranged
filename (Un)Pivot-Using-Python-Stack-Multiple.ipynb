```python
import pandas as pd
```

```python
test = pd.ExcelFile('Input_Multiple.xlsx')
test_df = pd.read_excel(test, header=[0,1])
```

```python
# Set the first col. as an index
test_df = test_df.set_index([('Header 1', 'Unnamed: 0_level_1'),
            ('Header 2', 'Unnamed: 1_level_1')])
```

```python
# Stack resets the data into a different format.
# Slightly modified-Combined as one row

test_df = pd.DataFrame(test_df.stack(0).stack(0)).reset_index()

print(test_df)
```

```python
test_df = test_df.rename(columns={
    ('Header 1', 'Unnamed: 0_level_1'): 'Country', 
    ('Header 2', 'Unnamed: 1_level_1') : 'Country_Idx', 
    u'level_1' : 'year',
    u'level_2' : 'type', 
    0 : 'Medals'})

test_df
```
